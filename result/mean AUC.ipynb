```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, auc

# 讀取資料
import pandas as pd
#df=pd.read_csv('./AUC.csv')

# 在各種『決策門檻』（decision threshold）下，計算 『真陽率』（True Positive Rate；TPR）與『假陽率』（False Positive Rate；FPR）
fpr, tpr, threshold = roc_curve(df_total['actual'], df_total['predict'])
fpr1, tpr1, threshold1 = roc_curve(df_1['actual'], df_1['predict'])
fpr2, tpr2, threshold1 = roc_curve(df_2['actual'], df_2['predict'])
fpr3, tpr3, threshold1 = roc_curve(df_3['actual'], df_3['predict'])
fpr4, tpr4, threshold1 = roc_curve(df_4['actual'], df_4['predict'])
fpr5, tpr5, threshold1 = roc_curve(df_5['actual'], df_5['predict'])






#print(fpr, tpr, threshold)

auc0 = auc(fpr, tpr)
auc1 = auc(fpr1, tpr1)
auc2 = auc(fpr2, tpr2)
auc3 = auc(fpr3, tpr3)
auc4 = auc(fpr4, tpr4)
auc5 = auc(fpr5, tpr5)



## Plot the result
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, color = 'red', label = 'Mean AUC = %0.2f' % auc0)
plt.plot(fpr1, tpr1, color = 'green', label = 'fold 1 AUC = %0.2f' % auc1,linestyle="--",linewidth=0.5)
plt.plot(fpr2, tpr2, color = 'orange', label = 'fold 2 AUC = %0.2f' % auc2,linestyle="--",linewidth=0.5)
plt.plot(fpr3, tpr3, color = 'blue', label = 'fold 3 AUC = %0.2f' % auc3,linestyle="--",linewidth=0.5)
plt.plot(fpr4, tpr4, color = 'cyan', label = 'fold 4 AUC = %0.2f' % auc4,linestyle="--",linewidth=0.5)
plt.plot(fpr5, tpr5, color = 'purple', label = 'fold 5 AUC = %0.2f' % auc5,linestyle="--",linewidth=0.5)


plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()  
```

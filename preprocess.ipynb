```python
#this spt is fit ROI with 2 lung

import nibabel as nib
from scipy import ndimage
import os
import nibabel as nib
from scipy import ndimage
import numpy as np



#read,process,save




def add_nifti_file(df):
    


    for i in range (0,len(df)):

#    for i in range (0,1):
        
        code_ph_c  = str(df['ID3'][i]) + "_ct.nii" 

        path_ph_c = "C:/Users/85298/AI/CT_image/mask_combine_fit/" +  code_ph_c


        img = process_scan(path_ph_c)        
        

#########################

#        img , name = shift_x (img)



        if df['Label'][i] == "PE" :
            path_save_16 = "C:/Users/85298/AI/CT_image/202304/224_160_48/ori/"+ code_ph_c + ".gz"
#            path_save_16 = "C:/Users/85298/AI/CT_image/test/" + name +"_" + code_ph_c + ".gz"                
#            path_save_16 = "C:/Users/85298/AI/CT_image/pe_combine_resize_" + str(desired_width) + "_" + str(desired_depth) +"/"  + file         
#            print ("PE")
        
        else :
            path_save_16 = "C:/Users/85298/AI/CT_image/202304/224_160_48/ori/"+ code_ph_c + ".gz"
#            path_save_16 = "C:/Users/85298/AI/CT_image/test/" + name +"_" + code_ph_c + ".gz"
#            path_save_16 = "C:/Users/85298/AI/CT_image/nr_combine_resize_" + str(desired_width) + "_" + str(desired_depth) +"/"  + file         
#            print ("nr")
      
#########################     


        img16 = nib.Nifti1Image(img, np.eye(4))

#        img16.get_data_dtype() == np.dtype(np.int16) 

        nib.save(img16, os.path.join(path_save_16))       

    
    
    
def process_scan(path):
    #Read and resize volume
    # Read scan
    volume = read_nifti_file(path)
    # Crop image
    
    # Normalize
    volume = normalize(volume)

    # Resize width, height and depth
    volume = resize_volume(volume) 
    
    

    return volume
    
    
    
def read_nifti_file(filepath):
    """Read and load volume"""
    # Read file
    scan = nib.load(filepath)
    

    scan = scan.get_fdata()

    return scan

def normalize(volume):
    #Normalize the volume
    min = -500
    max = 1500
    volume[volume < min] = min
    volume[volume > max] = max
    volume = (volume - min) / (max - min)
    volume = volume.astype("float32")
    return volume


def resize_volume(img):
    
    #Resize across z-axis
    # Set the desired depth
    desired_depth = 48
    desired_width = 224
    desired_height = 160
    
    # Get current depth
    current_depth = img.shape[-1]
    current_width = img.shape[0]
    current_height = img.shape[1]
    
    # Compute depth factor
    depth = current_depth / desired_depth
    width = current_width / desired_width
    height = current_height / desired_height
    depth_factor = 1 / depth
    width_factor = 1 / width
    height_factor = 1 / height

    # Resize across z-axis
    img = ndimage.zoom(img, (width_factor, height_factor, depth_factor), order=1)
        
    return img
```
